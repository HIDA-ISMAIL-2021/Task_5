```python
df2=df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].agg('mean', numeric_only=True).sort_values(by="2022 Population",ascending=False)
df3=df2.transpose()
```

```python
df.select_dtypes(include='float')
```

```python
df3.plot()
plt.xticks(rotation=45) 
```

```python
plt.figure(figsize=(15,10))
df3.plot(kind="bar")
plt.xticks(rotation=45) 
```

```python
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Histograms of all numeric columns')
plt.show()
```

```python
plt.figure(figsize=(10,8))
numeric_df = df.select_dtypes(include=['float64', 'int64']) 
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```

```python
sns.scatterplot(x='Area (km²)', y='2020 Population', data=df)
plt.title('Population vs Area')
plt.show()
```
